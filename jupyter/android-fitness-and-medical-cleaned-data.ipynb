```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
pd. set_option("display.max_columns", None) 
```

```python
path = './../android'
android_files = glob.glob(os.path.join(path , "*.csv"))
dfs = []
for filename in android_files:
    df = pd.read_csv(filename)
    dfs.append(df)

df = pd.concat(dfs, axis=0, ignore_index=True)
```
